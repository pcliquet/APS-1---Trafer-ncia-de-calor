# 1)
[1,5 pontos] Qual a temperatura na extremidade da aleta? Apresente um gráfico da distribuição de
temperatura ($T_{analitica}$) ao longo do comprimento da aleta. Use legenda, identifique os eixos e as
unidades usadas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
Tb = 100 + 273 # Transformação para Kelvin
Ta = 25 + 273 # Transformação para Kelvin
thetab = Tb - Ta 
```

```python
h = 25
k = 240 #
t = 20e-3 #
w = 240e-3 #
L = 200e-3 #
Atr = w * t #
P = 2 * (w + t) #
m = sqrt(h * P / k * Atr)
```

```python
def theta_thetab(x, h, P, k, Atr, L):
    m = (h * P / (k * Atr)) ** 0.5
    thetab = Tb - Ta
    ratio = (np.cosh(m * (L - x)) + (h / (m * k)) * np.sinh(m * (L - x))) / (np.cosh(m * L) + (h / (m * k)) * (np.sinh(m * L)))
    M = thetab * sqrt(h * P * k * Atr)
    qa = M * (np.sinh(m * L) + (h / (m * k)) * np.cosh(m * L)) / (np.cosh(m * L) + (h / (m * k)) * np.sinh(m * L))
    return ratio, qa
```

```python
theta_thetab(L, h, P, k, Atr, L)
```

## Vamos criar a lista de valores de x

```python
X = np.linspace(0, L, 100)
Y = np.array([theta_thetab(x, h, P, k, Atr, L)[0] for x in X])
YT = np.array(Y * thetab + Ta)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(211)
plt.plot(X, Y)
plt.xlabel('x(m)')
plt.ylabel(r'$\theta / \theta_b$')

plt.grid(True)

plt.subplot(212)
plt.plot(X, YT)
plt.xlabel('x(m)')
plt.ylabel(r'$T(x) (K)$')
plt.grid(True)

plt.show()
```

```python
print(f'1) R: A temperatura na extremidade da aleta será => {theta_thetab(L, h, P, k, Atr, L)[0] * thetab + Ta} K')
```

Sabendo da condição da aleta infinita, tem-se como solução a equação abaixo:

$$\displaylines{
m_{a}^2 = \frac{h\cdot P}{k_a\cdot A_{tr}}\to m_a = \sqrt{\frac{h\cdot P}{k_a \cdot A_{tr}}}\\
m_{b}^2 = \frac{h\cdot P}{k_b\cdot A_{tr}}\to m_b = \sqrt{\frac{h\cdot P}{k_b \cdot A_{tr}}}\\
\frac{\theta}{\theta_a} = e^{-m_ax}\\
ln(\frac{\theta}{\theta_a}) = -m_ax\\
ln(\frac{\theta}{\theta_b}) = -m_bx\\
x = -\frac{ln(\frac{\theta}{\theta_a})}{m_a} = -\frac{ln(\frac{\theta}{\theta_b})}{m_b}\\
\frac{ln(\frac{\theta}{\theta_a})}{m_a} = \frac{ln(\frac{\theta}{\theta_b})}{m_b}\\
\frac{m_b}{m_a} = \frac{ln(\frac{\theta}{\theta_b})}{ln(\frac{\theta}{\theta_a})}\\
\frac{\sqrt{\frac{h\cdot P}{k_b \cdot A_{tr}}}}{\sqrt{\frac{h\cdot P}{k_a \cdot A_{tr}}}} = \frac{ln(\frac{\theta}{\theta_b})}{ln(\frac{\theta}{\theta_a})}\\
\frac{\frac{h\cdot P}{k_b \cdot A_{tr}}}{\frac{h\cdot P}{k_a \cdot A_{tr}}} = \frac{ln^2(\frac{\theta}{\theta_b})}{ln^2(\frac{\theta}{\theta_a})}\\
\frac{\frac{1}{k_b}}{\frac{1}{k_a}} = \frac{ln^2(\frac{\theta}{\theta_b})}{ln^2(\frac{\theta}{\theta_a})}\\
\frac{k_a}{k_b} = \frac{ln^2(\frac{\theta}{\theta_b})}{ln^2(\frac{\theta}{\theta_a})}\\
k_a = \frac{ln^2(\frac{\theta}{\theta_b})}{ln^2(\frac{\theta}{\theta_a})}\cdot k_b
}$$
